```python
df_calc = df_all.copy()
df_new =pd.read_excel('city_all_7.xlsx')
```

```python
df_new['R&D经费占GDP比重(%)'] = (df_calc['R&D内部经费支出（万元）'] / df_calc['地区生产总值(万元)']) * 100
    # 科学支出占财政支出比重 = (科学支出 / 地方财政一般预算内支出) × 100
df_new['科学支出占财政支出比重(%)'] = (df_calc['科学支出(万元)'] / df_calc['地方财政一般预算内支出(万元)']) * 100
    # 教育支出占财政支出比重 = (教育支出 / 地方财政一般预算内支出) × 100
df_new['教育支出占财政支出比重(%)'] = (df_calc['教育支出(万元)'] / df_calc['地方财政一般预算内支出(万元)']) * 100
    # 货物出口额占GDP比重 = (货物出口额 / 地区生产总值) × 100
df_new['货物出口额占GDP比重(%)'] = (df_calc['货物出口额(万元)'] / df_calc['地区生产总值(万元)']) * 100

    # ========== 2. 人员/当量类指标 ==========
    # R&D人员全时当量：默认1人=1人年（有小时数可改为 工作小时数/2000）
df_new['R&D人员全时当量(人年)'] = df_calc['R&D人员（人）']

    # ========== 3. 直接复用原数据已有指标（统一列名标注） ==========
df_new['第三产业增加值占GDP比重(%)'] = df_calc['第三产业增加值占GDP比重(%)']
df_new['人均地区生产总值(元)'] = df_calc['人均地区生产总值(元)']
df_new['GDP增长率(%)'] = df_calc['地区生产总值增长率(%)']
df_new['第一产业增加值占GDP比重(%)'] = df_calc['第一产业增加值占GDP比重(%)']
df_new['第二产业增加值占GDP比重(%)'] = df_calc['第二产业增加值占GDP比重(%)']
df_new['每百人互联网宽带用户_常住人口口径'] = df_calc['每百人互联网宽带用户_常住人口口径']
df_new['年末金融机构贷款余额(万元)'] = df_calc['年末金融机构各项贷款余额(万元)']

    # ========== 4. 每万人类指标（常住人口单位：万人，直接相除） ==========
    # 每万人高新技术企业数 = 累计注册高科技企业数 / 常住人口数(万人)
df_new['每万人高新技术企业数(家/万人)'] = df_calc['累计注册高科技企业数'] / df_calc['常住人口数(万人)']
    # 每万人专利授权数 = 专利授权数(件) / 常住人口数(万人)
df_new['每万人专利授权数(件/万人)'] = df_calc['专利授权数(件)'] / df_calc['常住人口数(万人)']
    # 每万人在校大学生数 = 普通高等学校在校学生数(人) / 常住人口数(万人)
df_new['每万人在校大学生数(人/万人)'] = df_calc['普通高等学校在校学生数(人)'] / df_calc['常住人口数(万人)']
    # 每万人医院床位数 = 医院、卫生院床位数(张) / 常住人口数(万人)
df_new['每万人医院床位数(张/万人)'] = df_calc['医院、卫生院床位数(张)'] / df_calc['常住人口数(万人)']

    # ========== 5. 缺失指标（需补充数据后计算） ==========
    # 城镇登记失业率 = (城镇登记失业人数 / (城镇就业人数 + 城镇登记失业人数)) × 100
df_new['城镇登记失业率(%)'] = None  # 补充数据后取消注释计算
    # df_calc['城镇登记失业率(%)'] = (df_calc['城镇登记失业人数'] / (df_calc['城镇就业人数'] + df_calc['城镇登记失业人数'])) * 100

    # 保留指定小数位数
df_new = df_new.round(4)
print("✅ 所有指标计算完成！")
```
