```python

import numpy as np
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 定义三角模糊数参数（示例值）
a_l = 2      # 下限值（最小可能值）
a_M = 5      # 中值（隶属度为1的点）
a_u = 8      # 上限值（最大可能值）

# 创建x轴数据
x = np.linspace(a_l - 1, a_u + 1, 1000)

# 定义隶属函数
def triangular_mf(x, a_l, a_M, a_u):
    """三角模糊数隶属函数"""
    mu = np.zeros_like(x)
    
    # 左半部分：a_l ≤ x ≤ a_M
    left_mask = (x >= a_l) & (x <= a_M)
    mu[left_mask] = (x[left_mask] - a_l) / (a_M - a_l)
    
    # 右半部分：a_M ≤ x ≤ a_u
    right_mask = (x >= a_M) & (x <= a_u)
    mu[right_mask] = (x[right_mask] - a_u) / (a_M - a_u)
    
    # 其他区域保持为0
    return mu

# 计算隶属度
mu = triangular_mf(x, a_l, a_M, a_u)

# 创建图形
fig, ax = plt.subplots(figsize=(5, 3))

# 绘制隶属函数曲线
ax.plot(x, mu, 'b-', linewidth=2.5, label=r'$\mu_{\tilde{A}}(x)$')

# 填充曲线下方的区域
ax.fill_between(x, 0, mu, alpha=0.3, color='lightblue')

# 标记关键点
# a_l点
ax.plot(a_l, 0, 'ro', markersize=10)
ax.annotate(r'$a_l$ (下限值)', xy=(a_l, 0), xytext=(a_l-0.8, -0.15),
            fontsize=9, ha='center',
            arrowprops=dict(arrowstyle='->', color='red'))

# a_M点（顶点）
ax.plot(a_M, 1, 'ro', markersize=10)
ax.annotate(r'$a_M$ (中值, 隶属度=1)', xy=(a_M, 1), xytext=(a_M, 1.15),
            fontsize=9, ha='center',
            arrowprops=dict(arrowstyle='->', color='red'))

# a_u点
ax.plot(a_u, 0, 'ro', markersize=10)
ax.annotate(r'$a_u$ (上限值)', xy=(a_u, 0), xytext=(a_u+0.8, -0.15),
            fontsize=9, ha='center',
            arrowprops=dict(arrowstyle='->', color='red'))

# 添加虚线辅助线
ax.axhline(y=1, color='gray', linestyle='--', alpha=0.5, xmax=0.55)
ax.axvline(x=a_M, color='gray', linestyle='--', alpha=0.5, ymax=0.9)

# 添加左侧斜率标注
mid_left = (a_l + a_M) / 2
ax.annotate('', xy=(a_l, 0), xytext=(a_M, 1),
            arrowprops=dict(arrowstyle='<->', color='green', lw=1.5))
ax.text((a_l+a_M)/2 - 0.5, 0.5, r'$\frac{x-a_l}{a_M-a_l}$', 
        fontsize=10, color='green', ha='center')

# 添加右侧斜率标注
ax.annotate('', xy=(a_u, 0), xytext=(a_M, 1),
            arrowprops=dict(arrowstyle='<->', color='green', lw=1.5))
ax.text((a_M+a_u)/2 + 0.5, 0.5, r'$\frac{x-a_u}{a_M-a_u}$', 
        fontsize=10, color='green', ha='center')

# 设置坐标轴
ax.set_xlim(a_l - 1.5, a_u + 1.5)
ax.set_ylim(-0.3, 1.4)
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel(r'$\mu_{\tilde{A}}(x)$', fontsize=10)
ax.set_title('图2-1 三角模糊数隶属函数', fontsize=16, pad=20)

# # 添加网格
# ax.grid(True, alpha=0.3, linestyle=':')

# 添加隶属度范围标注
ax.text(a_M, 0.5, '隶属度\n范围', fontsize=10, ha='center', 
        bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.3))

# plt.tight_layout()
# plt.savefig('/mnt/kimi/output/triangular_fuzzy_number.png', dpi=150, bbox_inches='tight')
plt.show()
# print("三角模糊数图2-1已生成")
```

```python
def FuzzyLeastSquaresRegression(X, Y, K, ε):
    """输入:
    X : 样本特征矩阵 (m × (k+1))
    Y : 响应变量向量 (m × 1)
    K : 最大迭代次数
    ε : 收敛阈值

    # 输出:
    # b : 回归系数向量"""

    m ← 样本数量

    # Step1 初始化隶属度
    for i = 1 to m do
        u[i] ← 1
    end for

    # Step2 初始回归（普通最小二乘）
    b ← (XᵀX)⁻¹XᵀY

    for k = 1 to K do

        # Step3 计算预测值
        Y_hat ← X * b

        # Step4 计算误差
        for i = 1 to m do
            e[i] ← Y[i] - Y_hat[i]
        end for

        # Step5 更新隶属度
        for i = 1 to m do
            u[i] ← exp(-(e[i])²)
        end for

        # Step6 构造权重矩阵
        U ← diag(u)

        # Step7 加权最小二乘更新回归系数
        b_new ← (XᵀUX)⁻¹XᵀUY

        # Step8 收敛判断
        if ||b_new - b|| < ε then
            break
        end if

        b ← b_new

    end for

    return b
```
